```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

# Correct file path
data_path = Path(r"C:\Users\Administrator\OneDrive\Desktop\Attachment\Senior_Data_Scientist_Assessment_Data (1).xlsx")

# Load Excel file
xls = pd.ExcelFile(data_path)

# List sheet names
sheets = xls.sheet_names

# Load each sheet into a dictionary
tables = {s: xls.parse(s) for s in sheets}

# Print sheet name + shape
for k, v in tables.items():
    print(k, v.shape)
```

```python
def clean_df(df):
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]
    # normalize datetime if present:
    for c in df.columns:
        if 'date' in c.lower() or 'created_at' in c.lower() or 'updated_at' in c.lower():
            try:
                df[c] = pd.to_datetime(df[c], errors='coerce')
            except:
                pass
    return df

for k in list(tables.keys()):
    tables[k] = clean_df(tables[k])
```

```python
import pandas as pd

#  COPY TABLES
customers = tables['Customers'].copy()
accounts = tables['Accounts'].copy()
leads = tables['Leads'].copy()

#  RENAME CUSTOMER ID 
customers = customers.rename(columns={'id': 'customer_id'})

if 'customer_id' not in accounts.columns:
    for c in accounts.columns:
        if 'customer' in c.lower():
            accounts = accounts.rename(columns={c: 'customer_id'})
            break

# DETECT BALANCE COLUMN 
bal_cols = [c for c in accounts.columns if 
            'balance' in c.lower() or 
            'amount' in c.lower() or 
            'outstanding' in c.lower()]

bal_col = bal_cols[0] if bal_cols else None

# AGGREGATION DICTIONARY 
agg_dict = {'account_count': ('customer_id', 'count')}

if bal_col:
    agg_dict['total_balance'] = (bal_col, 'sum')
    agg_dict['avg_balance'] = (bal_col, 'mean')

if 'created_at' in accounts.columns:
    accounts['created_at'] = pd.to_datetime(accounts['created_at'], errors='coerce')
    agg_dict['first_account_date'] = ('created_at', 'min')

# GROUP BY CUSTOMER
cust_agg = accounts.groupby('customer_id').agg(
    **{k: pd.NamedAgg(column=v[0], aggfunc=v[1]) for k, v in agg_dict.items()}
).reset_index()

# MERGE WITH CUSTOMERS 
cust_features = customers.merge(cust_agg, on='customer_id', how='left')

# COMPUTE DAYS SINCE FIRST ACCOUNT 
today = pd.Timestamp.today()

if 'first_account_date' in cust_features.columns:
    cust_features['days_since_first_account'] = (
        today - cust_features['first_account_date']
    ).dt.days.fillna(-1)
else:
    cust_features['days_since_first_account'] = -1

# FILL MISSING BALANCE COLUMNS 
for col in ['account_count', 'total_balance', 'avg_balance']:
    if col not in cust_features.columns:
        cust_features[col] = 0
    else:
        cust_features[col] = cust_features[col].fillna(0)

# CONFIRM FINAL COLUMNS 
print(cust_features.columns.tolist())
print(cust_features[['account_count', 'total_balance', 'avg_balance', 'days_since_first_account']].head())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# pick numeric features and one-hot encode categorical later
num_cols = ['account_count','total_balance','avg_balance','days_since_first_account']
X = cust_features[num_cols].fillna(0).copy()

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=0.95))  # keep 95% variance
])
X_trans = pipeline.fit_transform(X)

# choose k using silhouette and elbow
sil_scores = {}
inertia = {}
for k in range(2,9):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_trans)
    sil = silhouette_score(X_trans, labels)
    sil_scores[k] = sil
    inertia[k] = km.inertia_

# pick best k by silhouette (highest)
best_k = max(sil_scores, key=sil_scores.get)
km_best = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cust_features['cluster'] = km_best.fit_predict(X_trans)
```

```python
# profile clusters
profile = cust_features.groupby('cluster')[num_cols].agg(['count','mean','median'])
display(profile)

# add categorical summaries
if 'lead_source' in cust_features.columns:
    lead_profile = cust_features.groupby('cluster')['lead_source'].apply(lambda s: s.value_counts().head(3).to_dict())
    print(lead_profile)
```

Clustering Result: K=2 was chosen (likely by Silhouette). Segments are highly imbalanced 95% vs 5%. Key Finding (Cluster 0): This is the Established Majority Segment Count=4750 defined by high average tenure  1071 days.Key Finding (Cluster 1): This is the Minority Segment Count=250 defined by negative/anomalous tenure Mean=-1.0. This cluster likely represents new or problematic accounts.Critical Data Issue: Features total_balance, avg_balance, and account_count showed zero variance approx 0 and did not contribute to the segmentation model.Action: Must fix the source calculation for balance/count metrics and the tenure anomaly before using the cluster profiles for business strategy. The current clustering is almost entirely based on one feature: account age.
